```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('IPL_Matches_2008-2020.csv')
df.head()
```

```python
x = ['Royal Challengers Bangalore', 'Kolkata Knight Riders', 'Kings XI Punjab', 'Rajasthan Royals', 'Mumbai Indians', 
   'Chennai Super Kings', 'Sunrisers Hyderabad', 'Delhi Capitals', 'Delhi Daredevils', 'Rising Pune Supergiant',
   'Gujarat Lions', 'Pune Warriors', 'Kochi Tuskers Kerala', 'Deccan Chargers', 'Rising Pune Supergiants']

y = ['RCB', 'KKR', 'KXIP', 'RR', 'MI', 'CSK', 'SRH', 'DC', 'DD', 'RPS', 'GL', 'PWI', 'KTK', 'DCH', 'RPS']
```

```python
df.replace(x, y, inplace = True)
df.head()
```

```python
matches_played = df['team1'].value_counts() + df['team2'].value_counts()
matches_played
```

```python
matches_won = df['winner'].value_counts()
matches_won
```

```python
win_percentage = (matches_won / matches_played) * 100
win_percentage
```

```python
plt.figure(figsize = (10, 6))
plt.xlabel('Teams')
plt.ylabel('Win Percentage')
plt.title('IPL Teams Win Percentage')
clr = ['blue']
sns.barplot(x = win_percentage.index, y = win_percentage, palette = clr)
```

```python
toss_sub = df.loc[df['toss_winner'] == df['winner']]
toss_sub.head()
```

```python
twinner = toss_sub['winner'].value_counts()
twinner
```

```python
total_toss_wins = df['toss_winner'].value_counts()
total_toss_wins
```

```python
toss_percentage_win = (twinner / total_toss_wins) * 100
toss_percentage_win
```

```python
plt.figure(figsize = (10, 6))
plt.xlabel('Teams')
plt.ylabel('Toss Win Percentage')
plt.title('IPL Teams Toss Win Percentage')
clr = ['orange']
sns.barplot(x = toss_percentage_win.index, y = toss_percentage_win, palette = clr)
```

```python
max_player_of_match = df['player_of_match'].value_counts()
max_player_of_match.head(10)
```

```python
runs_sub = df.loc[df['result'] == 'runs']
runs_sub.head()
```

```python
defend_win = runs_sub['winner'].value_counts()
defend_win
```

```python
defend_win_percentage = (defend_win / matches_won) * 100
defend_win_percentage
```

```python
plt.figure(figsize = (10, 6))
plt.xlabel('Teams')
plt.ylabel('Defending Win Percentage')
plt.title('IPL Teams Defending Win Percentage')
clr = ['red']
sns.barplot(x = defend_win_percentage.index, y = defend_win_percentage, palette = clr)
```
